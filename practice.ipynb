```python
import pandas as pd
import seaborn as sns
import numpy as np
from sklearn.linear_model import LinearRegression

import matplotlib.pyplot as plt

# Load the data
df = pd.read_json('anscombe.json')

# Create a 2x2 grid for plotting
fig, axs = plt.subplots(2, 2, figsize=(10, 10))

# Define the series and axes
series = ['I', 'II', 'III', 'IV']
axes = axs.flatten()

# Set the same x and y limits for all plots
x_limits = (4, 20)
y_limits = (2, 14)

# Plot each series
for i, s in enumerate(series):
    ax = axes[i]
    data = df[df['Series'] == s]
    
    # Scatter plot
    sns.scatterplot(x='X', y='Y', data=data, ax=ax, label='Data')
    
    # Linear fit
    X = data['X'].values.reshape(-1, 1)
    Y = data['Y'].values
    model = LinearRegression().fit(X, Y)
    Y_pred = model.predict(X)
    
    # Plot the linear fit
    ax.plot(data['X'], Y_pred, color='red', label='Fit')
    
    # Set titles and labels
    ax.set_title(f'Series {s}')
    ax.set_xlim(x_limits)
    ax.set_ylim(y_limits)
    ax.legend()

# Adjust layout
plt.tight_layout()
plt.show()
```
